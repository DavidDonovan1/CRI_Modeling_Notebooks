```python
states_provinces = cfeature.NaturalEarthFeature(
        category='cultural',
        name='admin_1_states_provinces_lines',
        scale='50m',
        facecolor='none')

for curYear in range(1981, 1982):
    
    NOAA_years = np.where((pd.DatetimeIndex(NOAA.time.data).year==1991))
    NOAA_short = NOAA.sst[(NOAA_years[0]),:,:]
    consecuitiveDays = NOAA_short.mean('time') * 0
    max_consecuitiveDays = consecuitiveDays


    for day in range(len(NOAA_short.time)):
        for x in range(len(NOAA_short[day].data)):
            for y in range(len(NOAA_short[day].data[x])):
                if (NOAA_short[day].data[x][y] > min_Temp - variance and NOAA_short[day].data[x][y] < max_Temp + variance ):
                    consecuitiveDays[x][y] += 1
                    if consecuitiveDays[x][y] > max_consecuitiveDays[x][y]:
                        max_consecuitiveDays[x][y] = consecuitiveDays[x][y]
                else:
                    consecuitiveDays[x][y] = 0
        
    
    fig = plt.figure(figsize=[20,10])
    
    Consecuitive_Plot = fig.add_subplot(1,1,1,projection=crs.LambertConformal())
    Consecuitive_Plot.set_title(str(curYear), fontsize=20)
    Consecuitive_Plot.set_extent([230,245,20,43], crs.Geodetic())
    Consecuitive_Colormesh = Consecuitive_Plot.pcolormesh(lon, lat, max_consecuitiveDays,
                 transform=crs.PlateCarree(),
                 cmap='jet',vmax=365, vmin=0)

    Consecuitive_Plot.add_feature(cfeature.COASTLINE)
    Consecuitive_Plot.add_feature(cfeature.LAND)
    Consecuitive_Plot.add_feature(cfeature.BORDERS)
    Consecuitive_Plot.add_feature(states_provinces, edgecolor='gray')
    Consecuitive_Plot.gridlines()

    colorbar = fig.colorbar(NOAA_Colormesh, shrink=0.7, anchor=('SW'))
    colorbar.set_label('Degrees Celcius',fontsize=24)
    colorbar.ax.tick_params(labelsize=22)
    fig.savefig(r"D:\CRI_Modelling_PC\CRI Maps\HAB 2020\Consecuitive Days Per Year\ " + str(curYear)+ ".png", bbox_inches='tight')
```
