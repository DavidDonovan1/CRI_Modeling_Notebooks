```python
import numpy as np
import pandas as pd
import xarray as xr
import seaborn as sns
import cartopy.crs as crs
import cartopy.feature as cfeature
import matplotlib
import matplotlib.pyplot as plt
```

```python
bathymetry = xr.open_dataset(r'D:\CRI_Modelling_PC\Bathymetry\crm.nc')
```

```python
finalBath = bathymetry.Band1
```

```python
lat = finalBath.lat.data
lon = finalBath.lon.data
```

```python
fig = plt.figure(figsize=[40,20])

plotBath = fig.add_subplot(1,1,1,projection=crs.PlateCarree())
plotBath.set_title('Bathymetry Spawning: 5 - 18 meters',fontsize=50)

levels = [-1000,-18,-5,0,1000]

c = ('powderblue', 'limegreen', 'powderblue','beige')

plot = plotBath.contourf(lon, lat, finalBath,levels,
             transform=crs.PlateCarree(),colors=c)

plotBath.gridlines()

colorbar = fig.colorbar(plot, shrink=0.7, anchor=('SW'))
colorbar.set_label('Depth (m)',fontsize=40)
colorbar.ax.tick_params(labelsize=30)

fig.savefig(r"D:\CRI_Modelling_PC\CRI Maps\HAB 2020\HalibutSpawningBath.jpeg", bbox_inches='tight')
```
